```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt   
%matplotlib inline
%pylab inline
pylab.rcParams['figure.figsize'] = (20, 10)  
```

```python
SPY = pd.read_csv('SPY.csv')
```

```python
SPY['Volatility'] = (SPY['High']-SPY['Low'])/SPY['Close']
```

```python
SPY['Volume_delta'] = SPY['Volume'] - SPY['Volume'].shift(1)
```

```python
SPY['Volatility_delta'] = SPY['Volatility'] - SPY['Volatility'].shift(1)
```

```python
SPY.plot(x="Volume_delta",y="Volatility_delta",style=".")
```
